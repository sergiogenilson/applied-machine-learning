```python
# Convert from ['person', 'cat', 'dog'] format to coco format

import glob

validation_dir = "../datasets/PP3b/labels/valid/"
train_dir = "../datasets/PP3b/labels/train/"

def convert_classes(dir):
    for file in glob.glob(dir + "*.txt"):
        new_labels = ""
        with open(file) as f:
            for line in f:
                s = line.split(" ", 1)
                if s[0] == "0":  # person
                    new_labels += "0 " + s[1]
                if s[0] == "1": # cat
                    new_labels += "15 " + s[1]
                elif s[0] == "2": # dog
                    new_labels += "16 " + s[1]
            f.close()
        
        fout = open(file, "w")
        fout.write(new_labels)
        fout.close()

convert_classes(validation_dir)
convert_classes(train_dir)
```
